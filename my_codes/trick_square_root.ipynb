## Trick to find Square root
$$ \sqrt{a} = \frac {a + b}{2* \sqrt{b} } $$

where $b$ is the integer which is closest to $a$ and is perfectly square.

eg-
$$ \sqrt{10} = \frac {10 + 9}{2* \sqrt{9} } = \frac {19}{6} $$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(0, 10001, 1)
```

```python
y_t = np.sqrt(x)
```

```python
y_t[27]
```

```python
y_c = []
for i in range(len(y_t)):
    l = math.floor(y_t[i])
    u = math.ceil(y_t[i])
    if (-l + y_t[i]) < (u - y_t[i]):
        x_ = (i+l**2) / (2*l)
        y_c.append(x_)
    elif (-l + y_t[i]) > (u - y_t[i]):
        x_ = (i+u**2)/(2*u)
        y_c.append(x_)
    else:
        y_c.append(y_t[i])
```

```python
plt.figure(figsize=(10,6),dpi=150)
plt.plot(x, y_t, label='True')
plt.plot(x, y_c, label='Calculated');
error = y_t - y_c
plt.plot(x, error, label='Error');
plt.legend();
```

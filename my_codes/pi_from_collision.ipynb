## Get the value of pi from the no. of elastic collisions of two objects

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
momentum = {}
velocity = {}
def add_to_dic(u1,u2,momentum,velocity):

    momentum1 = np.sqrt(m1)*u1
    momentum2 = np.sqrt(m2)*u2
    if momentum2 in momentum.keys():
        momentum[momentum2].append(momentum1)
        velocity[u2].append(u1)
    else:
        momentum[momentum2] = [momentum1]
        velocity[u2] = [u1]
```

```python
m1 = 1 ; u1 = 0 # innitially at rest

d = float(input("Enter the no. of digits:  "))
m2 = float(100**(d-1))       # d is the no. of digits of the value of pi
u2=1

v1 = v2 = 0   #final velocity

collision = 0
add_to_dic(u1,u2,momentum,velocity)
add_to_dic(u1,u2,momentum,velocity) 
```

```python
while True:
    collision +=1

    
    v2 = float( ( ( ( (m2 - m1)* u2 ) + 2*m1*u1 ) / (m1 + m2)) )
    v1 = float(( -( (m2 - m1)* u1 ) + 2*m2*u2 ) / (m1 + m2))   
    u2 = float(v2)
    u1 = float(v1)

    
    
    if(  u2 < 0 and abs(u1) <  abs(u2)):    # velocity of m2 is greater and is in backward direction
        break;

    add_to_dic(u1,u2,momentum,velocity)

    collision +=1
    u1 = -u1  #after colliding with wall

    if(   u2 < 0 and abs(u1) <  abs(u2)):    # velocity of m2 is greater and is in backward direction
        break;
    add_to_dic(u1,u2,momentum,velocity) 

```

```python
if u1>0:
    collision +=1   

#adding one collision for the case if u1 is less than u2 but u is moving towards the wall then that last collision with wall was not considered
```

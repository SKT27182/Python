### Removing the logo

```python
texts_to_remove = [

    '''"___\\n",''',

    '../Pierian_Data_Logo.png',
    'http://www.pieriandata.com',
    'www.pieriandata.com',
    'For more information, visit us at',
    'Copyright Pierian Data',
    'https://www.udemy.com/user/joseportilla/',
    'Copyright by Pierian Data Inc.',
    'Content Copyright by Pierian Data'
]
```

```python
import os
for root, _ , files in os.walk('.'):
    for file in files:
        if file.endswith('.ipynb'):
            with open(os.path.join(root, file), 'r') as f:
                text = f.read()
                for keys in texts_to_remove:
                    text = text.replace(keys, '')
                with open(os.path.join(root, file), 'w') as f:
                    f.write(text)
```

### Converting the notebooks to HTML

```python
from html_setup.notebook import Notebook
import os
f = Notebook('.')
```

```python
outputs = []
for folder, _, files in os.walk('.'):
    for file in files:
        if file.endswith('.ipynb'):
            file_path = os.path.join(folder, file)
            if ".ipynb_checkpoints" in file_path:
                continue
            if "__pycache__" in file_path:
                continue
            
            print(f"Converting {file}")
            try:
                output = f.export_notebook(file_path)
                outputs.append(output)
            except Exception as e:
                print(f"Error converting {file}")
                print(e)
```

```python
string = ''
present_dir = 'Jupyter Notebooks Guide'

container = """<div>{heading}
{content}
</div>
"""

txt = """
<li>
    <a href="Python-Bootcamp_Jose_Portilla/{final_path}" target = 'blank' class = 'links'>{title}</a>
</li>
"""

container_ = ''

for file_path in outputs:
    final_path = file_path  #'/'.join(file_path.split("\\")[1:])
    if ".ipynb_checkpoints" in final_path:
        continue
    elif "__pycache__" in file_path:
        continue
    elif ('Exercise.html' in file_path) or ('Exercises.html' in file_path)  or ('Assessment.html' in file_path) :
        continue
    else:
        title = final_path.split("/")[-1]
        title = " ".join(title.split("-")[1:])
        title=title.split(".")[0]
        title =re.sub(r"(\w)([A-Z])", r"\1 \2", title)
        if present_dir == current_dir(final_path):
            txt_ = txt.format(final_path=final_path, title=title)
            string += txt_
            #print(string)
        else:
            container_ += container.format(heading=present_dir,content=string)
            string = ''
            txt_ = txt.format(final_path=final_path, title=title)
            string += txt_
            present_dir = current_dir(final_path)

container_ += container.format(heading=present_dir,content=string)

print(container_)
```
